```python
import matplotlib.pyplot as plt
import numpy as np

labels = {}
with open('scrapy_test/spiders/Output/labels.txt', 'r', encoding='utf-8') as f:
        for line in f.readlines():
            labels[line.strip()] = 0
with open('scrapy_test/spiders/Output/_trainSet.txt', 'r', encoding='utf-8') as f:
        for para in f.readlines():
            category = para.split('\t', 1)[0]
            labels[category] += 1

#plt.xticks(np.arange(0, 120, step=10))
plt.grid(True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
plt.barh(list(labels.keys()), labels.values(), alpha=0.5)
plt.title('Training set', loc='left')
plt.xlabel('Article counts')
```

```python
for key in labels:
    labels[key] = 0
with open('scrapy_test/spiders/Output/testAns.txt', 'r', encoding='utf-8') as f:
        for para in f.readlines():
            category = para.strip()
            labels[category] += 1

plt.xticks(np.arange(0, 120, step=10))
plt.grid(True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.2)
plt.barh(list(labels.keys()), labels.values(), alpha=0.5)
plt.title('Test set', loc='left')
plt.xlabel('Article counts')
```
